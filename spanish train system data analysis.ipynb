```python
import numpy as np 
import pandas as pd 
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../input/renfe.csv', index_col=0)
data.head()
```

**Finding total number of null values in the dataset**

```python
data.isnull().sum()
```

**Replace NA values in price column with the mean price in price column**

```python
data['price'].fillna(data['price'].mean(),inplace=True)
```

**Dropping the column which wouldn't be used**

```python
data.drop('insert_date',axis=1,inplace=True)
```

**Dropping the rows containing Null values in the columns train_class and fare**

```python
data.dropna(inplace=True)
```

**Now checking the number of na values**

```python
data.isnull().sum()
```

**No. of trains running in Spain**

```python
fig,ax = plt.subplots(figsize=(20,10))
ax = sns.countplot(data['train_type'])
plt.show()
```

**No. of people boarding from each station**

```python
fig,ax = plt.subplots(figsize=(20,10))
ax = sns.countplot(data['origin'])
plt.show()
```

**No. of people getting off at each station**

```python
fig,ax = plt.subplots(figsize=(20,10))
ax = sns.countplot(data['destination'])
plt.show()
```

**No. of trains of each class**

```python
fig,ax = plt.subplots(figsize=(20,10))
ax = sns.countplot(data['train_class'])
plt.show()
```

**No. of tickets bought from each category**

```python
fig,ax = plt.subplots(figsize=(20,10))
ax = sns.countplot(data['fare'])
plt.show()
```

**violin plot of train type vs price**

```python
f,ax = plt.subplots(figsize=(18,8))
ax = sns.violinplot(x='train_type',y='price',data=data)
plt.show()
```

**Violin plot of train class vs price**

```python
f,ax = plt.subplots(figsize=(18,8))
ax = sns.violinplot(x='train_class',y='price',data=data)
plt.show()
```
